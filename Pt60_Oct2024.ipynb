```python
%matplotlib widget
import HF_cursor_function as hfcp
import matplotlib.pyplot as plt
```

```python
#new aluminum
sample_filename = "data4/"+"al_103-303-10042024-1.txt"
sample_filename1 = "data4/"+"sapphire_103-303-10042024-1.txt"
sample_filename2 = "data4/"+"empty_103-303-10042024-1.txt"
sample_composition = "Al"
points_list_1,points_list_2,points_list_3,weight1,weight2,sample_data = hfcp.hf_cursor1(sample_filename,sample_filename1,sample_filename2,sample_composition)
```

```python
cp_samp_al,T_samp_al = hfcp.cp(points_list_1,points_list_2,points_list_3,sample_composition,weight1,weight2,sample_data)
```

```python
T_samp_al_lit,cp_samp_al_lit = [500, 600, 700, 800, 900],[26.84, 27.89, 29.10, 30.56, 32.31]
errors = [0.01 * cp for cp in cp_samp_al_lit]
plt.figure()
plt.title("heat capacity of Al")
plt.errorbar(T_samp_al_lit,cp_samp_al_lit,yerr=errors,fmt="b--o",ecolor="b",label="Literature Value Al cp")
plt.scatter(T_samp_al+273.15,cp_samp_al,c="r",label="measured Al cp")
plt.ylabel("cp(J/mol.K)")
plt.xlabel("temperature(K)")
plt.legend()
plt.show()
print(T_samp_al+273.15,cp_samp_al)
```
